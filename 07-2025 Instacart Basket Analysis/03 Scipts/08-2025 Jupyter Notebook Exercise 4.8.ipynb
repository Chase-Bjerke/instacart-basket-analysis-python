## 04 - Customer Loyalty Flags

```python
df_ords_prods_merge.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

```python
# 3 step process to create a max column: groupby/transform/create
df_ords_prods_merge['max_order'] = df_ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Check work, ensure max_order column was created
df_ords_prods_merge.head(15)
```

Max Column created

```python
# Assign loyalty flags based on max_order values --- Loyal Cust.
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
# Assign loyalty flags based on max_order values --- Reg. Cust.
df_ords_prods_merge.loc[(df_ords_prods_merge['max_order'] <= 40) & (df_ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
# Assign loyalty flags based on max_order values --- New Cust.
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Check work, ensure Flags worked
df_ords_prods_merge.head(15)
```

## 06 - Classifying Users by Spending Behavior

```python
# Find user AVG price per product
user_avg_price = df_ords_prods_merge.groupby('user_id')['prices'].mean().reset_index()
```

```python
# New Avg Price column to keep things clean
user_avg_price['avg_price'] = user_avg_price['prices']
```

```python
# work check
user_avg_price.head()
```

```python
# Function to create spending flags per customer
def spending_category(x):
    if x < 10:
        return 'Low spender'
    else:
        return 'High spender'
```

```python
# Create a new spending flags column and apply spending flags to customers
user_avg_price['spending_flag'] = user_avg_price['avg_price'].apply(spending_category)
```

```python
# Work check
user_avg_price.head()
```

## 07 - Frequent vs Non Frequent Customers

```python
# Group by user_id and calculate the median days between orders
user_medians = df_ords_prods_merge.groupby('user_id')['days_since_prior_order'].median().reset_index()
```

```python
# Keep track of user medians
user_medians.rename(columns={'days_since_prior_order': 'median_days'}, inplace=True)
```

```python
user_medians.head()
```

Looks like the column was created and renamed correctly

```python
# Define order freq. flag function
def order_frequency_flag(median):
    if median > 20:
        return 'Non-frequent customer'
    elif 10 <= median <= 20:
        return 'Regular customer'
    else:
        return 'Frequent customer'
```

```python
# Apply function
user_medians['order_frequency_flag'] = user_medians['median_days'].apply(order_frequency_flag)
```

```python
# Work check
user_medians.head()
```

```python
# Merge this work to the main dataset
df_ords_prods_merge = df_ords_prods_merge.merge(
    user_medians[['user_id', 'order_frequency_flag']],
    on='user_id',
    how='left'
)
```

```python
# Merge and work check
df_ords_prods_merge.head()
```

Looks like the merge was successful

# 08 Work on Final Report

#### Price Range

```python
# Define price range labeling function
def price_label(row):
    if row['prices'] <= 5:
        return 'Low-range product'
    elif 5 < row['prices'] <= 15:
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High-range product'
    else:
        return 'Not enough data'
```

```python
# Apply function to full dataset
df_ords_prods_merge['price_range'] = df_ords_prods_merge.apply(price_label, axis=1)
```

```python
# Quick preview
print(df_ords_prods_merge[['prices', 'price_range']].head(10))
```

```python
# Count of each price range
price_counts = df_ords_prods_merge['price_range'].value_counts(dropna=False)
```

```python
# Percentage of each price range
price_percent = df_ords_prods_merge['price_range'].value_counts(normalize=True, dropna=False) * 100
```

```python
# Combine into summary DataFrame
price_summary = pd.DataFrame({
    'Price Range': price_counts.index,
    'Count': price_counts.values,
    '% of Total Orders': price_percent.round(1).astype(str) + '%'
})
```

```python
# Display summary
print(price_summary)
```
