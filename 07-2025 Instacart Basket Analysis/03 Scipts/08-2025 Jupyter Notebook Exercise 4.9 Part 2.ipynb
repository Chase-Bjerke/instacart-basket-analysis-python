## Controlled Merge (note: only run once or if dataset gets messed up)

```python
# Compare and check columns, I want all non-duplicated columns from cust
cust_cols = df_ords_prods_merge_cust.columns.tolist()
main_cols = df_ords_prods_merge.columns.tolist()
```

```python
# Keep only columns from cust that aren't already in main
non_duplicate_cols = [col for col in cust_cols if col not in main_cols]
```

```python
# Define merge keys that uniquely identify each row (order-product pair)
merge_keys = ['order_id', 'product_id']
```

```python
# Subset the customer file to only the new columns + merge keys
cust_subset = df_ords_prods_merge_cust[merge_keys + non_duplicate_cols]
```

```python
# Merge the subset into the main file using a left join to preserve all rows
df_enriched = df_ords_prods_merge.merge(cust_subset, on=merge_keys, how='left')
```

```python
# Step 6: Confirm the shape and inspect the final column list
print(df_enriched.shape)
print(df_enriched.columns.tolist())
```

```python
# Export to save progress
df_final_v2 = df_enriched
df_final_v2.to_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'df_final_v2.pkl'))
```

```python
# Group by hour and calculate average spend
avg_spend_by_hour = df_ords_prods_merge.groupby('order_hour_of_day')['prices'].mean()

# Plot it
fig, ax = plt.subplots(figsize=(10, 6))
avg_spend_by_hour.plot(kind='bar', color='mediumseagreen', ax=ax)

# Labeling
ax.set_title('Average Spend per Order by Hour of Day', fontsize=14)
ax.set_xlabel('Hour of Day (0â€“23)', fontsize=12)
ax.set_ylabel('Average Spend ($)', fontsize=12)
ax.set_xticks(range(0, 24))
ax.grid(axis='y', linestyle='--', alpha=0.7)

# Insight below chart
plt.figtext(0.5, 0.01, "Higher average spend observed during late morning and early afternoon hours.", ha='center', fontsize=10)

# Export
plt.tight_layout()
fig.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'avg_spend_by_hour.png'), dpi=300)
plt.show()
```
